
# Distribution Plots

```python
#To visualize distribution in a dataset
```

```python
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#Dist plot
#Univariate set of observations
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
#Joint plot
#basically combine two Distplots for 
#Bivariate data
#we also have a "Kind" parameter here
#which allows us to choose how we wanna compare the two
#variables
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
#Pair plot
#Basically a jointplot for every single numerical col
#in your dataset
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
#Rug plot
#univariate
#Concept is you just draw a simple dash for every data point
#across the distribution line
#kde plot -- kernel density estimation plot

sns.rugplot(tips['total_bill'])
```

```python
#KDE plot
sns.kdeplot(tips['total_bill'])
```
